```python
import pandas as pd
import numpy as np
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.cm as cm
import matplotlib.pyplot as plt
# % matplotlib inline
```

```python
messages = pd.read_csv('new_spam.csv', encoding='latin-1')
```

```python
msg = messages['message'].tolist()
single_msg = ' '.join(msg)
```

```python
# All data
wordcloud_spam = WordCloud(background_color="white", stopwords = STOPWORDS).generate(single_msg)

# Lines 2 - 5
plt.figure(figsize = (9,10))
plt.imshow(wordcloud_spam, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
spam_msgs = []
ham_msgs = []
for data in messages.values:
    if data[0] == 'spam':
        spam_msgs.append(data[1])
    elif data[0] == 'ham':
        ham_msgs.append(data[1])
all_spam = ' '.join(spam_msgs)
all_ham = ' '.join(ham_msgs)
```

```python
# spam data word cloud
wordcloud_spam = WordCloud(background_color="white", stopwords = STOPWORDS).generate(all_spam)

# Lines 2 - 5
plt.figure(figsize = (9,10))
plt.imshow(wordcloud_spam, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
# ham data word cloud
wordcloud_spam = WordCloud(background_color="white", stopwords = STOPWORDS).generate(all_ham)

# Lines 2 - 5
plt.figure(figsize = (9,10))
plt.imshow(wordcloud_spam, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
messages['length'] = messages['message'].apply(len)
plts = messages.hist(column='length',by='class',bins=50, figsize=(15,6), color="grey")
for axis in plts.flatten():
    axis.set_xlabel("Length of the message")
    axis.set_ylabel("Number of messages")
    axis.set_title("")
```
